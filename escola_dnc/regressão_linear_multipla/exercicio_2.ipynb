```python
import pandas as pd
import numpy as np
import seaborn as sns

import statsmodels.formula.api as smf
```

```python
df = pd.read_excel('C+¦pia de Limite_Credito_Dummy.xlsx')
df.head(10)
```

```python
df = df.rename(columns={'Gênero': 'Região',
                        'Região': 'Gênero'})
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
df = pd.get_dummies(data=df, prefix=['Escolaridade', 'Estado', 'Genero'])
```

```python
df.corr()
```

```python
# Montando o modelo 
function = 'LimitedoChequeEspecial~Idade+RendimentoTotal+Salario+LimitedeCreditoImediato+Escolaridade_Fundamental_Medio+Escolaridade_Superior_Pos+Estado_RJ+Estado_SP+Genero_F+Genero_M'

model = smf.ols(formula=function, data=df).fit()
print(model.summary())
```

```python
# Parâmetros de previsão 
new_params = {
        'Salario': [15000],
        'Estado_RJ': [False],
        'Estado_SP': [True],
        'Genero_F': [False],
        'Genero_M': [True]
}

# Predizendo cheque especial
predict = model.predict(new_params)
```
