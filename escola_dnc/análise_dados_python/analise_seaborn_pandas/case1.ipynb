```python
# importanto bibliotecas
import pandas as pd
import seaborn as sns
```

```python
# Lendo a base de dados em excel
aluguel_casa = pd.read_excel('houses_rent.xlsx')
```

```python
# Mostrando head da tabela
aluguel_casa.head(10)
```

```python
# Mostrando os últimos itens da lista
aluguel_casa.tail(10)
```

```python
# Atualizando os nomes das colunas
aluguel_casa = aluguel_casa.rename(columns={'parking spaces': 'parking_spaces',
                                             'rent amount': 'rent_amount',
                                               'property tax': 'property_tax', 
                                               'fire insurance': 
                                               'fire_insurance'})
```

```python
# Visualização dos dados
aluguel_casa.head(10)
```

```python
# Identificando o tipo dos dados
aluguel_casa.dtypes
```

```python
# Alterando tipos dos dados
aluguel_casa = aluguel_casa.astype({'rent_amount': 'float',
                                     'property_tax': 'float', 
                                     'fire_insurance': 'float', 
                                     'total': 'float'})
aluguel_casa.dtypes
```

```python
# Visualização dos dados
aluguel_casa.head(10)
```

```python
# Informações adicionais sobre a tabela
aluguel_casa.info()
```

```python
# Descrevendo algumas colunas da tabela
aluguel_casa[['area', 'rooms', 'bathroom', 'parking_spaces', 'rent_amount', 
              'property_tax', 'fire_insurance', 'total']].describe()
```

```python
# Filtrando tabela
aluguel_casa = aluguel_casa[(aluguel_casa['total']<=4000)]
aluguel_casa.head()
```

```python
# Mais tipos de filtros
aluguel_casa = aluguel_casa[(aluguel_casa['total'] <=4000) & (aluguel_casa['animal'] == 'accept') & (aluguel_casa['rooms'] == 2) | (aluguel_casa['rooms'] == 3 )]
aluguel_casa.head()
```

```python
# Descrevendo novamente a tabela
aluguel_casa.describe()
```

```python
# Filtrando com query
aluguel_casa = aluguel_casa.query("total <= 4000 and animal == 'accept'")
aluguel_casa.head()
```

```python
# Mais filtros para a tabela
aluguel_casa = aluguel_casa.query(("total <= 4000 and animal == 'accept' and rooms == 2 or rooms == 3"))
aluguel_casa.head()
```

```python
# Descrevendo a tabela
aluguel_casa.describe()
```

```python
# Agrupando dados
agg_city_price = aluguel_casa.groupby(['city'])['total'].agg('mean').reset_index()
agg_city_price
```

```python
# Outro agrupamento
agg_bath_room_price = aluguel_casa.groupby(['bathroom', 'rooms'])['total'].agg('mean').reset_index()
agg_bath_room_price
```

```python
# Gráfico de barras para qtd de quartos em relação ao preço
ax = sns.barplot(data=agg_bath_room_price, x='rooms', y='total', color = 'green');
ax.set(xlabel = 'qtd quartos', ylabel = 'media do preço');
```

```python
# Gráfico de barras, relação de banheiros com preço
ax = sns.barplot(data=agg_bath_room_price, x='bathroom', y='total', color = 'red');
ax.set(xlabel='qtd banheiros', ylabel='media do preço');
```

```python
# Agrupando e filtrando dados
agg_area = aluguel_casa[['area']]
agg_area = agg_area.query('area <= 100')
```

```python
# Descrição da nova tabela
agg_area.describe()
```

```python
# Histograma da tabela agrupada
hist = sns.histplot(data=agg_area, x='area', color='brown');
hist.set(xlabel='Area', ylabel='Contagem de casas');
```

```python
# Outro histograma da base
sns.histplot(data=agg_area, x= "area", color = "red");
```

```python
# Histograma com a contagem das cidades em relação com preço e cidade
city_hist = sns.histplot(data=aluguel_casa, x='total', hue='city');
city_hist.set(xlabel='Preço', ylabel='Contagem de Casas');
```

```python
# Descrição da tabela
aluguel_casa.describe()
```

```python
# Filtrando por andar e quantidade de apartamentos
aluguel_casa.groupby('floor').index.nunique().sort_values(ascending=False)
```

```python
# Agrupando
agg_floor_price = aluguel_casa.groupby('floor')["total"].agg('mean').reset_index()
agg_floor_price
```

```python
# Gráfico de barras dos dados acima
sns.barplot(data=agg_floor_price, x='floor', y='total', color='yellow').set(
    xlabel='Andar', ylabel='Preço'
);
```
